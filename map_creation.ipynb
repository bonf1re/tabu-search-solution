```python
import numpy as np
from random import randint
from random import randrange
import gc
```

```python
height = 30
width = 60
```

```python
# lil plotting lol
def plotting(my_map,height,width):
    for i in range(0,height):
        for j in range(0,width):
            if my_map[i][j]==0:
                print('0', end='')
            else:
                print('_',end='')
        print()
    return
```

```python
# creating array
def creation():
    my_map = []
    for i in range(0, height):
        column = []
        for j in range(0, width):
            column.append(0)
        my_map.append(column)
    return my_map
```

```python
def stands_insertion(my_map,height,width):
    # our stands will have 2 thickness, bc each product will occupy a coordinate
    # will use way in and way out in order to introduce it
    # Horizontally
    thickness = 2
    check_radius= 2
    for n in range(0,200):
        print(n)
        longitude = randrange(10,21,5) # random number from 10 to 20 with step of 5
        # direction of creation will be foward downward
        # horizontally
        init_cord = [randrange(check_radius,width-1-check_radius,2),randint(check_radius,height-1-thickness-check_radius)] 
        #print(init_cord)
        #print(longitude)
        can_fit = True
        for i in range(init_cord[0]-check_radius-1,init_cord[0]+longitude+check_radius,check_radius):
            for j in range(init_cord[1]-check_radius-1,init_cord[1]+thickness+check_radius,check_radius):
                try:
                    if my_map[j][i] == 1:
                        can_fit = False
                except:
                    can_fit= False
        if can_fit == True:
            for i in range(init_cord[0],init_cord[0]+longitude):
                for j in range(init_cord[1],init_cord[1]+thickness):
                    my_map[j][i] = 1
                    #print(my_map[j][i])
        if can_fit == False:
            init_cord = [randrange(check_radius,width-1-check_radius,2),randint(check_radius-1,height-1-thickness-check_radius)] 
            can_fit=True
            for i in range(init_cord[0]-check_radius-1,init_cord[0]+thickness+check_radius,check_radius):
                for j in range(init_cord[1]-check_radius-1,init_cord[1]+longitude+check_radius,check_radius):
                    try:
                        if my_map[j][i] == 1:
                            can_fit = False
                    except:
                        can_fit= False
            if can_fit == True:
                for i in range(init_cord[0],init_cord[0]+thickness):
                    for j in range(init_cord[1],init_cord[1]+longitude):
                        my_map[j][i] = 1
    return my_map
    
```

```python
def usable_coordinates(my_map,height,width):
    counter = 50
    coordinates =[]
    for i in range(0,height):
        for j in range(0, width):
            will_pick = randint(0,2)
            if will_pick == 1 and counter > 0:
                coordinates.append([i,j])
                counter = counter - 1
    return coordinates
            
```

```python
for n in range(0,50):
    my_map =creation()
    #plotting(my_map,height,width)
    my_map = stands_insertion(my_map,height,width)
    plotting(my_map,height,width)
    coordinates=usable_coordinates(my_map,height,width)
    thefile = open('map_'+str(n)+'.txt', 'w')
    for i in range(0,height):
        for j in range(0,width):
            thefile.write("%s" % str(my_map[i][j]))
        thefile.write("\n")
    thefile.close()
    del thefile
    gc.collect()
    thefile2 = open('coordinates_'+str(n)+'.txt', 'w')
    for item in coordinates:
        thefile2.write("[%s,%s]; " % (str(item[0]),str(item[1])))
    thefile2.close()
    del thefile2
    del my_map
    gc.collect()
```
